### Pclass - Embarked relation

```python
df = pd.read_csv('data/train.csv')
# df[df['Pclass']==3]['Embarked']
transfer_dict = {'S' : 0, 'Q' : 1, 'C' : 2, np.NAN : -1}

for i in range(1,4):
    pclass_em = df[df['Pclass']==i]['Embarked'].map(transfer_dict)
    sqc = [0, 0, 0, 0]
    for irow in pclass_em.index:
        val = pclass_em[irow]
        sqc[val] += 1
    plt.bar(['S', 'Q', 'C', 'NaN'], sqc)
    plt.title('Pclass_{}'.format(i))
    plt.legend()
    plt.show()
```
